### Importando as bibliotecas Numpy e Matplotlib

```python
from matplotlib import pyplot as plt #Biblioteca do Python para criação de gráficos
import numpy as np
%matplotlib inline
```

### Plotando o 1° gráfico
Gráfico simples

```python
x,  y = [3,6,9,12,15],[2,4,6,8,10] #Valores de X e Y respectivamente para o gráfico linha
plt.plot(x, y, c = "black") #Gráfico estilo linha
plt.scatter(x, y, c = "red") #Gráfico estilo pontos

plt.title("Título") #Título do gráfico
plt.xlabel("Eixo X") #Título para a variável X
plt.ylabel("Eixo Y") #Título para a variável y
plt.grid(True) #coloca grades no gráfico

plt.show() #Plota o gráfico
```

### Plotando o 2° gráfico

Gráfico com legenda

```python
x = np.linspace(-10, 10, 500) #Valores de amostra

plt.plot(x,np.sin(x), c = "green", label="sin") #Plota o gráfico seno
plt.plot(x, np.cos(x), c = "blue", label = "cos") #Plota o gráfico cosseno
plt.grid(True) #Insere a grade
plt.legend() #Insere a legenda 
plt.show() #Plota o gráfico
```

### Gráfico de uma equação matemática



```python
def ex2(x):
    """
    Função
    """
    return 4*x**2 - 7*x +5

def diff_ex2(x):
    """
    Derivada da função
    """
    return 8*x - 7

#####################################################
# valores de amostra 
#####################################################

x = np.linspace(-10, 10, 500) 
y = ex2(x)

#####################################################
# plotando o gráfico
#####################################################

plt.plot(x, y, label="f(x)")
plt.plot(x, diff_ex2(x), label="df/dx")

#####################################################
# definindo o ponto mínimo
#####################################################

minimoX = 7/8 
minimoY = ex2(minimoX)

#####################################################
# exibindo o ponto 
#####################################################

plt.scatter(minimoX, minimoY, c="red") 

plt.annotate('Mínimo', # anotação
ha = 'center', va = 'bottom', # formatação da anotação
xytext = (minimoX-5, minimoY+200), #posição do texto
xy = (minimoX, minimoY), #posição inicial da seta
arrowprops = { 'facecolor' : 'yellow', 'shrink' : 0.02 }) #formatação da seta

plt.title("$4x^2-7x+5$") #título do gráfico
plt.legend() # insere a legenda
plt.grid(True) #insere a grade
plt.show() #exibe o gráfico
```
