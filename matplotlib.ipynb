# Matplotlib

```python
from matplotlib import pyplot as plt #Biblioteca do Python para criação de gráficos
import numpy as np
%matplotlib inline
```

```python
x,  y = [3,6,9,12,15],[2,4,6,8,10]
plt.plot(x, y, c = "black") #Valores de X e Y respectivamente para o gráfico linha
plt.scatter(x, y, c = "red") #Valores de X e Y respectivamente para o gráfico pontos
##formatação
plt.title("Título") #Título do gráfico
plt.xlabel("Eixo X") #Título para a variável X
plt.ylabel("Eixo Y") #Título para a variável y
plt.grid(True) #coloca grades no gráfico

plt.show() #Plota o gráfico
```

# Anotações

```python
def ex2(x):
    """
    Função
    """
    return 4*x**2 - 7*x +5

def diff_ex2(x):
    """
    Derivada da função
    """
    return 8*x - 7

#####################################################
# valores de amostra 
#####################################################

x = np.linspace(-10, 10, 500) 
y = ex2(x)

#####################################################
# plotando o gráfico
#####################################################

plt.plot(x, y, label="f(x)")
plt.plot(x, diff_ex2(x), label="df/dx")

#####################################################
# definindo o ponto mínimo
#####################################################

minimoX = 7/8 
minimoY = ex2(minimoX)

#####################################################
# exibindo o ponto 
#####################################################

plt.scatter(minimoX, minimoY, c="red") 

plt.annotate('Mínimo', # anotação
ha = 'center', va = 'bottom', # formatação da anotação
xytext = (minimoX-5, minimoY+200), #posição do texto
xy = (minimoX, minimoY), #posição inicial da seta
arrowprops = { 'facecolor' : 'black', 'shrink' : 0.05 }) #formatação da seta

plt.title("$4x^2-7x+5$") #título do gráfico
plt.legend() # insere legenda
plt.grid(True) #grade
plt.show() #exibe o gráfico
```
